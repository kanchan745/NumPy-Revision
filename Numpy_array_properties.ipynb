#How to check shape , size and type of array

```python
import numpy as np
```

**Shape of an array.**

```python
#check how many rows and column in an array : give int no.

#use it when we 've multi dimension data

arr_2d = np.array([[1,2,3],[4,5,6]])
print(arr_2d.shape)

#give output(2-rows , 3- columns)
```

# Size attribute : give total no. of elements

```python
#use when we 've huge data
arr = np.array([[10,20,30],[40,50,60]])
print(arr.size)
```

# ndim : attribute with the help , we can check no. of dimensions


```python

arr_1d = np.array( [1,2,3])
arr_2d = np.array([[5,6,7],[4,5,6]])
arr_3d = np.array([[[8,9],[10,11],[6,12],[5,75]]])

print(arr_1d.ndim)
print(arr_2d.ndim)
print(arr_3d.ndim)

```

```python
arr = np.array([10,20,30.5,40])
#to check the type of data
print(arr.dtype)
```

#conversion of data inside array


```python
#arr.astype()
arr = np.array([1.2,2.5,3.8])
print(arr.dtype)
int_arr = arr.astype(int)
print(int_arr.dtype)
print(int_arr)
```

Mathematical operation : without using loops



```python
arr =np.array([10,20,30])
print(arr +5)
print(arr *2)
print(arr**2)
```

**Aggregation function**

#To find summary or summarise data.  ex : average products from data

np.sum(array) -> add all elements
np.mean(array) -> calculate average
np.min(array) -> calculate smallest value in an array
np.std(array) -> To calculate standard deviation
np.var(array) -> To calculate variance
These are the some functions...






```python
arr = np.array([10,20,30,40,50])
print(np.sum(arr))
print(np.mean(arr))
print(np.min(arr))
print(np.std(arr))
print(np.var(arr))
```
